Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
data[ data['country'].str.contains('est', case=False) ]['country'].unique()
```

```python
data[(data.quantity > 0) 
     & ((data.country == 'Lithuania')|(data.country == 'Latvia')|(data.country == 'Estonia')) 
     & (data.year >= 2005) & (data.year <= 2010)
     & ((data.category == 4) | (data.category == 21) | (data.category == 12))
    ].quantity.sum()
```

```python
# with isin
data[(data.country.isin(['Lithuania','Latvia', 'Estonia'] )) &
     (data.year.isin([2005,2006, 2007,2008,2009,2010] )) &
     (data.category.isin([4,12, 21] )) &
     (data.quantity > 0)
    ].quantity.sum()
```
